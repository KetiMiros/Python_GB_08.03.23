```python
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_train.csv')
```

```python
df.head()
```

```python
df[(df['population'] > 500) ]
```

```python
print(df['median_house_value'].max())
print(df['median_house_value'].min())
print(df['median_house_value'].mean())
print(df['median_house_value'].median())

```

```python
df['median_house_value'].max(), df['median_house_value'].min(), df['median_house_value'].mean().round(2), df['median_house_value'].median()
```

```python
import seaborn as sns
sns.scatterplot(data=df, y="population", x="housing_median_age", hue="median_house_value")
sns.relplot(x="population", y="median_house_value", kind="line", data=df)
```

```python
sns.relplot(x = 'population', y = 'median_house_value', kind = 'line', data = df)
     
```

```python
sns.histplot(data=df, x="median_house_value")
#Столбчатая диаграмма. 
#По оси x указано значение, а по оси y - встречаемость(кол-во таких значений в выборке).
     
```

```python
df.loc[df['population'] <= 500, 'pop_group'] = 'Жителей до 500' #создание группы до 500 жителей
df.groupby('pop_group')['median_house_value'].min().plot(kind='bar') #минимальная среднее значение 
```

```python
#Максимальная цена в группе до 500 чел.
df.groupby('pop_group')['median_house_value'].max().plot(kind='bar')
```

```python
df.groupby('pop_group')['median_house_value'].mean().plot(kind='bar') 
```

```python
df.groupby('pop_group')['median_house_value'].median().plot(kind='bar') # среднее значение
```
